#### Differences of Interest Across Categories

Categories to Check Across KPIs:
- ['Branch', 'City', 'Customer type', 'Gender', 'Product line', 'Payment']

KPIs:
- ['Unit price', 'Quantity', 'Total', 'cogs', 'gross income', 'Rating']

```python
line = alt.Chart(walmart_df).mark_line().encode(
    x = 'Date:T',
    y = 'mean(gross income)',
    color='Customer type'
).properties(width = 500, height = 250).facet('Product line:N', columns=1)

line
```

```python
bar = alt.Chart(walmart_df).mark_bar().encode(
    y = 'Customer type',
    x = 'sum(gross income)',
    color='Customer type'
).properties(width = 200, height = 100).facet('Product line:N', columns=1)

bar
```

```python
bar2 = alt.Chart(walmart_df).mark_bar().encode(
    x = alt.X('Product line:N', sort='y'),
    y = 'sum(gross income):Q',
    color='Product line:N'
).properties(width = 250, height = 200)# .facet('Customer type', columns=1)

bar2
```

```python
bar2 = alt.Chart(walmart_df).mark_bar().encode(
    x = alt.X('Customer type:N', sort='y'),
    y = 'sum(gross income):Q',
    color='Product line:N'
).properties(width = 250, height = 200)# .facet('Customer type', columns=1)

bar2
```

```python
bar2 = alt.Chart(walmart_df).mark_bar().encode(
    x = alt.X('Branch:N', sort='y'),
    y = 'sum(gross income):Q',
    color='Product line:N'
).properties(width = 250, height = 200)# .facet('Customer type', columns=1)

bar2
```

```python
walmart_df_mean = walmart_df.copy()
walmart_df_mean = walmart_df_mean.groupby(['Date','Branch']).mean(numeric_only=True)
print(walmart_df_mean)

# walmart_df_mean['Branch'] = walmart_df['Branch']

alt.Chart(walmart_df_mean).mark_area().encode(
    x='Date:T',
    y=alt.Y('gross income:Q', stack='normalize'),
    color='Branch:N'
)#.facet('City')

# walmart_df_mean['Branch']
# walmart_df['Branch']
```
